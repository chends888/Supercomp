```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.system('./vector_O0')
os.system('./vector_O1')
# os.system('./vector_O2')
os.system('./vector_O3')
os.system('./vector_O3_flags')
```

```python
loglist = []
pow3list = []
pow3multlist = []
sumlist = []
sqrtlist = []

with open('O0_times.json') as O0:
    with open('O1_times.json') as O1:
#         with open('O2_times.json') as O2:
        with open('O3_times.json') as O3:
            with open('O3flags_times.json') as O3flags:
                o0data = json.load(O0)
                o1data = json.load(O1)
#                 o2data = json.load(O2)
                o3data = json.load(O3)
                o3fdata = json.load(O3flags)
                for i in range(8):
                    loglist.append([o0data[['log'][0]][i], o1data[['log'][0]][i], o3data[['log'][0]][i], o3fdata[['log'][0]][i]])
                    pow3list.append([o0data[['pow3'][0]][i], o1data[['pow3'][0]][i], o3data[['pow3'][0]][i], o3fdata[['pow3'][0]][i]])
                    pow3multlist.append([o0data[['pow3mult'][0]][i], o1data[['pow3mult'][0]][i], o3data[['pow3mult'][0]][i], o3fdata[['pow3mult'][0]][i]])
                    sumlist.append([o0data[['sum'][0]][i], o1data[['sum'][0]][i], o3data[['sum'][0]][i], o3fdata[['sum'][0]][i]])
                    sqrtlist.append([o0data[['sqrt'][0]][i], o1data[['sqrt'][0]][i], o3data[['sqrt'][0]][i], o3fdata[['sqrt'][0]][i]])
                
```

```python
# Log
groups = loglist
group_labels = ['10', '100', '1000', '10000', '100000', '1000000', '10000000', '20000000']

# Convert data to pandas DataFrame.
df = pd.DataFrame(groups, index=group_labels).T

# Plot.
pd.concat(
    [df.loc[0].rename('O0'), df.loc[1].rename('O1'), 
     df.loc[2].rename('O3'), df.loc[3].rename('O3 with flags')],axis=1).plot.bar()
plt.xlabel('time(s)', fontsize=13)
plt.ylabel('vector size', fontsize=13)
plt.title('Log', fontsize=16)
```
